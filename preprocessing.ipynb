```python
#import necessary libraries
import pandas as pd
import json
```

```python
with open('match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
df_match.isna().sum()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
with open('batting.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting.isna().sum()
```

```python
df_batting['dismissal'].fillna('', inplace=True)

# Apply the function while checking for NaN values
df_batting['out/not_out'] = df_batting['dismissal'].apply(lambda x: "out" if (isinstance(x, str) and len(x) > 0) else "not_out")
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.loc[df_batting.SR=="-","SR"]=0
df_batting.loc[df_batting.runs=="-",["runs","balls","4s","6s"]]=1
```

```python
with open('bowling.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

```python
with open('player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)
df_players.drop(columns=["input"],inplace=True)
print(df_players.shape)
df_players.head(10)
```

```python
import pandas as pd
import difflib

# Load the CSV files into DataFrames
df_names = df_players # DataFrame with names
df_images = pd.read_csv('/content/players_image.csv')  # DataFrame with images

# Function to find approximate matches between names
def find_matching_name(name, names_list):
    matches = difflib.get_close_matches(name, names_list, n=1, cutoff=0.7)
    return matches[0] if matches else None

# Create a mapping between names in both DataFrames
name_mapping = {}
for name in df_names['name']:
    matching_name = find_matching_name(name, df_images['name'])
    if matching_name:
        name_mapping[name] = matching_name

# Map the names in df_names to names in df_images and get corresponding image URLs
df_names['image'] = df_names['name'].map(name_mapping)
df_names['image'] = df_names['image'].map(df_images.set_index('name')['image'])

# Save the updated names DataFrame with image URLs to a new CSV file or use it as needed
df_names.to_csv('names_with_images.csv', index=False)
```
